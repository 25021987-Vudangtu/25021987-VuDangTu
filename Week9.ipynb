```python
#BAI1
def merge_sort_rec(a,b):
    if not a:
        return b
    if not b:
        return a
    if a[0]<=b[0]:
        return [a[0]]+merge_sort_rec(a[1:],b)
    else:
        return [b[0]]+merge_sort_rec(b[1:],a)
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(merge_sort_rec(a,b))

```

```python
#BAI2
def count1(matrix,mid):
    n,m=len(matrix),len(matrix[0])
    row,col=0,m-1
    count=0
    while row<n and col>=0:
        if matrix[row][col]>=mid:
            count+= n-row
            col-=1
        else:
            row+=1
def binary_search(low,high,matrix,k):
    if low==high:
        return low
    mid=(low+high)//2
    count=count1(matrix,mid)
    if count>=k:
        return binary_search(mid,high,matrix,k)
    else:
        return binary_search(low,mid-1,matrix,k)
a=list(map(int,input().split()))
b=list(map(int,input().split()))
c=list(map(int,input().split()))
matrix=[]
matrix.append(a)
matrix.append(b)
matrix.append(c)
k=int(input())
print(binary_search(matrix[0][0],matrix[len(matrix)-1][len(matrix)-1],matrix,k))
```

```python
#BAI3
def solve(s,index):
    if index>=len(s)-1:
        return s
    min=s[index]
    target=-1
    for i in range(len(s)-1,index,-1):
        if s[i]<min:
            min=s[i]
            target=i
    if target!=-1:
        s[index],s[target]=s[target],s[index]
        return s
    return solve(s,index+1)
s=input()
chars=list(s)
a=solve(chars,0)
result="".join(a)
print(int(result))
```

```python
#BAI4
def solve(a,b):
    if len(b)==0:
        return a.islower() or a==" "
    if len(a)==0:
        return False
    chara=a[0]
    charb=b[0]
    if chara.isupper():
        if chara == charb:
            return solve(a[1:],b[1:])
        else:
            return False
    else:
        delete=solve(a[1:],b)
        save=False
        if chara.upper()==charb:
            save=solve(a[1:],b[1:])
        return delete or save
n=int(input())
for i in range(n):
    a=input()
    b=input()
    if solve(a,b):print("YES")
    else:print("NO")
```

```python
#BAI6
def solve():
    n=int(input())
    num=list(map(int,input().split()))
    i=n-2
    while i>=0 and num[i]>num[i+1]:
        i-=1
    if i>=0:
        j=n-1
        while num[j]<=num[i]:
            j-=1
        num[i],num[j]=num[j],num[i]
    left=i+1
    right=n-1
    while left<right:
        num[left],num[right]=num[right],num[left]
        left+=1
        right-=1
    print(*num)
solve()
```

```python
#bai7
def canPlaceFlowers(flowerbed, k):
    if k==0:
        return True
    n=len(flowerbed)
    for i in range(n):
        if flowerbed[i]==0:
            left=(i==0)or flowerbed[i-1]==0
            right=(i==n-1) or flowerbed[i+1]==0
            if right and left:
                flowerbed[i]=1
                k-=1
                if k==0:
                    return True
    return False
flowerbed=list(map(int,input().split()))
k=int(input())
print(canPlaceFlowers(flowerbed,k))
```

```python
#BAI8
def is_path_crossing(moves):
    x,y=0,0
    save={(0,0)}
    for moves in moves:
        if moves =="U":y+=1
        elif moves =="D":y-=1
        elif moves =="R":x+=1
        elif moves =="L":x-=1
        if (x,y) in save:
            return True
        save.add((x,y))
    return False
moves=input()
print(is_path_crossing(moves))
```

```python
#BAI9
def Median(nums):
    if nums[0]>nums[1]:
        nums[0],nums[1]=nums[1],nums[0]
    if nums[2]>nums[3]:
        nums[2],nums[3]=nums[3],nums[2]
    if nums[0]>nums[2]:
        nums[0],nums[2]=nums[2],nums[0]
        nums[1],nums[3]=nums[3],nums[1]
    if nums[1]>nums[4]:
        nums[1],nums[4]=nums[4],nums[1]
    if nums[1]>nums[2]:
        nums[1],nums[2]=nums[2],nums[1]
        nums[4],nums[3]=nums[3],nums[4]
    if nums[2]<nums[4]:
        return nums[2]
    else:
        return nums[4]
nums=list(map(int,input().split()))
print(Median(nums))
```

```python
#BAI10
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    radius=0
    j=0
    for houses in houses:
        while j<len(heaters)-1 and abs(heaters[j+1]-houses)<=abs(heaters[j]-houses):
            j+=1
        dist=abs(houses-heaters[j])
        radius=max(radius,dist)
    return radius
houses=list(map(int,input().split()))
heaters=list(map(int,input().split()))
print(findRadius(houses,heaters))
```
