```python
#BAI3
def tongbp(x):
  s=0
  while x>0:
    s+=(x%10)**2
    x=x//10
  return s
try:
  a=int(input())
  check=set()
  while a!=1 and (a not in check):
    check.add(a)
    a=tongbp(a)
  if a==1:
    print("Yes")
  else:
    print("No")
except ValueError:
    print("No")








```
