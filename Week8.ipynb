```python
#BAI1
class Cylinder:
  def __init__(self,r,l):
    self.r=r
    self.l=l
  def dientich(self):
    S=(2*3.14*(r**2))+(2*3.14*r)*l
    return S
  def thetich(self):
    V=3.14*r**2*l
    return V
r,l=map(int,input().split())
C1=Cylinder(r,l)
print(f"{C1.dientich():.2f}")
print(f"{C1.thetich():.2f}")
```

```python
#BAI6
class Rectange:
  def __init__(self,w,h,k):
    self.w=w
    self.h=h
    self.k=k
  def dtich(self):
    S=self.w*self.h
    return S
  def chuvi(self):
    C=2*(self.w+self.h)
    return C
  def phongto(self):
    print(f"{self.dtich()*self.k**2:.2f} {self.chuvi()*self.k:.2f}")

w,h,k=map(float,input().split())
rec=Rectange(w,h,k)
rec.phongto()
```
