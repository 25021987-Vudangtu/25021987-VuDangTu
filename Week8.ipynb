```python
#BAI1
class Cylinder:
  def __init__(self,r,l):
    self.r=r
    self.l=l
  def dientich(self):
    S=(2*3.14*(r**2))+(2*3.14*r)*l
    return S
  def thetich(self):
    V=3.14*r**2*l
    return V
r,l=map(int,input().split())
C1=Cylinder(r,l)
print(f"{C1.dientich():.2f}")
print(f"{C1.thetich():.2f}")
```
