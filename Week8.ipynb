```python
#BAI1
class Cylinder:
  def __init__(self,r,l):
    self.r=r
    self.l=l
  def dientich(self):
    S=(2*3.14*(r**2))+(2*3.14*r)*l
    return S
  def thetich(self):
    V=3.14*r**2*l
    return V
r,l=map(int,input().split())
C1=Cylinder(r,l)
print(f"{C1.dientich():.2f}")
print(f"{C1.thetich():.2f}")
```

```python
#BAI2
class Date:
  def __init__(self, day, month, year):
    self.day = day
    self.month = month
    self.year = year

  def is_leap_year(self):
    if self.year % 400 == 0 or (self.year % 4 == 0 and self.year % 100 != 0):
      return True
    else:
      return False

  def is_valid(self):
    if self.month < 1 or self.month > 12:
      return False
    days_in_month = [31, 29 if self.is_leap_year() else 28,\
                     31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    if self.day < 1 or self.day > days_in_month[self.month - 1]:
      return False
    return True

  def next_date(self):
    days_in_month = [31, 29 if self.is_leap_year() else 28,\
                     31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    self.day += 1
    if self.day > days_in_month[self.month - 1]:
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1

  def format(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year}"


try:
    d, m, y = map(int, input().split('/'))
    date = Date(d, m, y)

    if not date.is_valid():
        print("INVALID")
    else:
        date.next_date()
        print(date.format())

except:
    print("INVALID")
```

```python
#BAI5
class ShoppingCart:
  def __init__(self):
    self.items = []

  def add_item(self, name, price):
    product ={'name': name, 'price': price}
    self.items.append(product)

  def remove_item(self, name):
    for item in self.items:
      if item['name'] == name:
        self.items.remove(item)
        return


  def if_empty(self):
    if len(self.items) == 0:
      print('cart is empty')
    else:
      print('cart is not empty')

  def total_price(self):
    s = 0
    for item in self.items:
      s += item['price']
    return s

  def show_cart(self):
    for item in self.items:
      print(item['name'], item['price'])

  def clear(self):
    self.items = []


cart1 = ShoppingCart()
  #main
while True:
    print("\n===== MENU =====")
    print("2. Thêm vào giỏ hàng một sản phẩm mới")
    print("3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên để xóa)")
    print("4. Kiểm tra xem giỏ hàng có rỗng không")
    print("5. Tính tổng tiền giỏ hàng")
    print("6. Hiển thị toàn bộ sản phẩm trong giỏ")
    print("7. Xóa toàn bộ giỏ hàng")

    choice = input("Chọn 2-7: ")

    if choice == 2:
      name = input('nhap ten san pham: ')
      price = float(input('nhap gia san pham: '))
      cart1.add_item(name, price)
    if choice == 3:
      name = input('nhap ten san pham can xoa: ')
      cart1.remove_item(name)
    if choice == 4:
      cart1.if_empty()
    if choice == 5:
      print(cart1.total_price())
    if choice == 6:
      cart1.show_cart()
    if choice == 7:
      cart1.clear()

    stop = input("Bạn có muốn thoát? ")
    if stop in ["yes", "YES", "Yes"]:
      break
    else:
      continue

```

```python
#BAI6
class Rectange:
  def __init__(self,w,h,k):
    self.w=w
    self.h=h
    self.k=k
  def dtich(self):
    S=self.w*self.h
    return S
  def chuvi(self):
    C=2*(self.w+self.h)
    return C
  def phongto(self):
    print(f"{self.dtich()*self.k**2:.2f} {self.chuvi()*self.k:.2f}")

w,h,k=map(float,input().split())
rec=Rectange(w,h,k)
rec.phongto()
```

```python
#BAI7
class Fraction():
    def __init__(self,num,den):
        if den==0:
            raise ValueError("Mẫu số không thể bằng 0")
        if den<0 :
            num=-num
            den=-den
        common=self.ucln(num,den)
        self.num=num//common
        self.den=den//common
    def ucln(self,a,b):
        a,b= abs(a) ,abs(b)
        while b!=0:
            a,b=b,a%b
        return a
    def add(self, other):
        new_num = self.num * other.den + self.den * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
    def sub(self, other):
        new_num = self.num * other.den - self.den * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
    def mul(self,other):
        new_num=self.num*other.num
        new_den=self.den*other.den
        return Fraction(new_num,new_den)
    def div(self,other):
        new_num=self.num*other.den
        new_den=self.den*other.num
        return Fraction(new_num,new_den)
    def __str__(self):
        return f"{self.num}/{self.den}"
input_data=input().split()
if len(input_data) == 5:
    a=int(input_data[0])
    b=int(input_data[1])
    op=input_data[2]
    c=int(input_data[3])
    d=int(input_data[4])
    f1=Fraction(a,b)
    f2=Fraction(c,d)
    result=None
    if op=="+":
        result=f1.add(f2)
    elif op=="-":
        result=f1.sub(f2)
    elif op=="*":
        result=f1.mul(f2)
    elif op=="/":
        result=f1.div(f2)
    if result:
        print(result)

```

```python
#BAI8
class BankAccount():
    def __init__(self,owner,blance):
        self.owner=owner
        self.blance=float(blance)
    def deposit(self,amount):
        self.blance+=amount
    def withdraw(self,amount):
        if amount<=self.blance:
            self.blance-=amount
    def __str__(self):
        return f"{self.owner}:{self.blance}"
Data=input().split()
a=Data[0]
b=float(Data[1])
my_account=BankAccount(a,b)
n=int(input())
for _ in range(n):
    command=input().split()
    read_cmd=command[0]
    amount=float(command[1])
    if read_cmd == "DEPOSIT":
        my_account.deposit(amount)
    elif read_cmd =="WITHDRAW":
        my_account.withdraw(amount)
print(f"{my_account.blance:.2f}")
```

```python
#BAI9
class Student:
    def __init__(self,name):
        self.name=name
        self.score=[]
    def add_score(self,subject,score):
        self.score.append(score)
    def avg(self):
        if not self.score:
            return 0.0
        else:
            return sum(self.score)/len(self.score)
    def rank(self):
        average=self.avg()
        if average >=8:
            return "Excellent"
        elif 6.5<=average<8:
            return "Good"
        elif 5<=average<6.5:
            return "Average"
        else:
            return "Poor"
name=input()
n=int(input())
student=Student(name)
for _ in range(n):
    data=input().split()
    subject=data[0]
    score=float(data[1])
    student.add_score(subject,score)
average_score=student.avg()
print(f"{student.name} {average_score:.2f} {student.rank()}")
```

```python
#BAI10
class Book:
    def __init__(self,title, author, year):
        self.title=title
        self.author=author
        self.year=year
class Library:
    def __init__(self):
        self.book=[]
    def add(self,book):
        self.book.append(book)
    def count_by_author(self,name):
        count_name=0
        for book in self.book:
            if book.author==name:
                count_name+=1
        return count_name
    def find_by_year(self,year):
        count_year=0
        for book in self.book:
            if int(book.year)==int(year):
                count_year+=1
        return count_year
n=int(input())
library=Library()
for _ in range(n):
    input_data=input().split()
    command=input_data[0]
    if command=="ADD":
        a=" ".join(input_data[1:])
        b=a.split(';')
        if len(b)==3:
            title=b[0].strip()
            author=b[1].strip()
            year=b[2].strip()
            library.add(Book(title,author,year))
    elif command=="COUNT":
        author_name=" ".join(input_data[1:]).strip()
        print(library.count_by_author(author_name))
    elif command=="COUNTYEAR":
        year=" ".join(input_data[1]).strip()
        print(library.find_by_year(year))
```

```python
#BAI12
class ComplexNumber:
    def __init__(self,realPart,imaginaryPart):
        self.realPart=float(realPart)
        self.imaginaryPart=float(imaginaryPart)
    def multiply(self,other):
        new_realPart=self.realPart*other.realPart-self.imaginaryPart*other.imaginaryPart
        new_imaginary=self.realPart*other.imaginaryPart+self.imaginaryPart*other.realPart
        return ComplexNumber(new_realPart,new_imaginary)
line1=input().split()
if line1:
    c1=ComplexNumber(line1[0],line1[1])
    line2=input().split()
    if line2:
        c2=ComplexNumber(line2[0],line2[1])
        result=c1.multiply(c2)
        print(f"{result.realPart} {result.imaginaryPart}")
```

```python
#BAI13
class Triangle:
    def __init__(self,edge1:int,edge2:int,edge3:int):
        self.edge1=edge1
        self.edge2=edge2
        self.edge3=edge3
    def check(self):
        a,b,c=self.edge1,self.edge2,self.edge3
        return (a + b > c) and (a + c > b) and (b + c > a)
    def getArea(self):
        if not self.check():
            return -1
        a,b,c=self.edge1,self.edge2,self.edge3
        p=(a+b+c)/2
        return (p*(p-a)*(p-b)*(p-c))**0.5
data=input().split()
if len(data)==3:
    e1,e2,e3=map(int,data)
    triangle=Triangle(e1,e2,e3)
    result=triangle.getArea()
    if result == -1:
        print("invalid")
    else:
        print(f"{result:.2f}")
```
