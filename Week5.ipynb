```python
#BAI1
def Maxoftwo(a,b):
  return max(a,b)
a,b=map(int,input().split())
print(Maxoftwo(a,b))
```

```python
#BAI2
def swap(a,b):
  a,b=b,a
  return a,b
x,y=map(int,input().split())
x,y=swap(x,y)

print(x,y)
```

```python
#BAI3
def prime(x):
  if x<2:return False

  for i in range(2,int(x**(1/2))+1):
    if x%i==0: return False
  return True
a=int(input())
if prime(a):
  print("True")
else:
  print("False")
```

```python
#BAI8
a,b=map(int,input().split())
def ham(a,b):
  a=bin(a)[2:]
  b=bin(b)[2:]
  while len(a)>len(b):
    b="0"+b
  while len(a)<len(b):
    a="0"+a
  s=0
  for i in range(0,len(a)):
    if a[i]!=b[i]:
      s+=1
  return s
print(ham(a,b))
```

```python
#BAI10
def anhxa(a,b):
  map1={}
  map2={}
  for i in range(0,len(a)):
    if a[i] in map1:
      if map1[a[i]]!=b[i]:return False

    else: map1[a[i]]=b[i]
    if b[i] in map2:
      if map1[a[i]]!=b[i]:return False

    else:map2[b[i]]=a[i]
  return True
a,b=input().split()
print(anhxa(a,b))
```
