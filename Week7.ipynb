```python
#BAI2
lst=list(map(int,input().split()))
x=int(input())
s=0
for i in lst:
  if i==x:
    s+=1
print(s)
```

```python
#BAI5
lst=list(map(int,input().split()))
X=int(input())
s=0
for i in range(0,len(lst)):
  for j in range(i+1,len(lst)):
    if lst[i]+lst[j]==X:
      s+=1
print(s)
```

```python
#BAI6
a=eval(input())
lst=[]
for i in range(len(a)):
  if i%2==0:
    for j in a[i]:
      lst.append(j)
  else:
    a[i]=a[i][::-1]
    for j in a[i]:
      lst.append(j)
print(lst)
```

```python
#BAI7
a=eval(input())

cnt=[]
for i in range(len(a)):
  s=1
  max1=0
  for j in range(i+1,len(a)):
    if a[i]<a[j] and max1<a[j]:
      max1=a[j]
      s+=1
  cnt.append(s)
b=max(cnt)
print(b)




```

```python
#BAI9
a=eval(input())
lst=[]
for i in range(len(a)):
  s=0
  for j in range(i+1,len(a)):
    if a[j]<a[i]:
      s+=1
  lst.append(s)
print(lst)
```

```python
#BAI11
lst=input().split()
w=input()
a=w+","
b=w+"."
print(lst.count(w)+lst.count(a)+lst.count(b))
```

```python
#BAI12
lst=[]
dapan=['A','B','C','D']
for i in range(4):
  s=input().strip()
  lst.append([s,dapan[i]])
lst.sort(key=lambda x:x[0])
for i in lst:
  print(i[1],end=" ")
```
