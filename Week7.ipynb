```python
#BAI1
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

```python
#BAI3
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        # Tìm phần tử nhỏ nhất trong phần danh sách chưa được sắp xếp
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # Đặt phần tử nhỏ nhất vào đúng vị trí
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

```python
#BAI8
def interval(lst,qu):
  n=[]
  lst.sort(key= lambda x : x[1]-x[0]+1)
  for q in qu:
    temp=-1
    for l,r in lst:
      if l <= q <= r:
        temp = r - l +1
        break
    n.append(temp)
  return n
lst = eval(input())
qu = eval(input())
print(interval(lst,qu))
```

```python
#BAI10
lst = eval(input())
n = len(lst)
m = int(input())
best = 0

for i in range(n):
  for j in range(i+1,n+1):
    s = sum(lst[i:j]) % m
    best = max(best,s)
print(best)
```
