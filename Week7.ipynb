```python
#BAI2
lst=list(map(int,input().split()))
x=int(input())
s=0
for i in lst:
  if i==x:
    s+=1
print(s)
```

```python
#BAI5
lst=list(map(int,input().split()))
X=int(input())
s=0
for i in range(0,len(lst)):
  for j in range(i+1,len(lst)):
    if lst[i]+lst[j]==X:
      s+=1
print(s)
```

```python
#BAI6
a=eval(input())
lst=[]
for i in range(len(a)):
  if i%2==0:
    for j in a[i]:
      lst.append(j)
  else:
    a[i]=a[i][::-1]
    for j in a[i]:
      lst.append(j)
print(lst)
```
