#Bai 7


```python
def prime(n):
    if (n < 2):
        return 0
    for i in range(2, int(n ** 0.5) + 1):
        if (n % i == 0):
            return 0
    return 1
n = int(input())
res = 0
for i in range(1, int(n ** 0.5) + 1):
    if (n % i == 0):
        res = max(res, i * prime(i), n // i * prime(n // i))
print(res)
```

#Bai 8


```python
def rev(n):
    x = 0
    while (n > 0):
        x = x * 10 + n % 10
        n //= 10
    return x
n = abs(int(input()))
c = 0
rn = rev(n)
while (n != rn):
    c += 1
    n += rn
    rn = rev(n)
print(c, n)
```

#Bai 9


```python
def f(x):
    d = [0] * 10
    while (x > 0):
        if (d[x % 10]):
            return 0
        d[x % 10] = 1
        x //= 10
    return 1
n = int(input())
n = int(n ** 0.5)
a = [i * i for i in range(1, n + 1) if f(i * i)]
print(a if len(a) else "No Number")
```
