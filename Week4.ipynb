#Bai 7


```python
def prime(n):
    if (n < 2):
        return 0
    for i in range(2, int(n ** 0.5) + 1):
        if (n % i == 0):
            return 0
    return 1
n = int(input())
res = 0
for i in range(1, int(n ** 0.5) + 1):
    if (n % i == 0):
        res = max(res, i * prime(i), n // i * prime(n // i))
print(res)
```

#Bai 8


```python
def rev(n):
    x = 0
    while (n > 0):
        x = x * 10 + n % 10
        n //= 10
    return x
n = abs(int(input()))
c = 0
rn = rev(n)
while (n != rn):
    c += 1
    n += rn
    rn = rev(n)
print(c, n)
```

```python
def f(x):
    d = [0] * 10
    while (x > 0):
        if (d[x % 10]):
            return 0
        d[x % 10] = 1
        x //= 10
    return 1
n = int(input())
n = int(n ** 0.5)
a = [i * i for i in range(1, n + 1) if f(i * i)]
print(a if len(a) else "No Number")
```

```python
#bai11
def count_even_divisors(n):
    count = 0
    for i in range(1, n + 1):
        if n % i == 0 and i % 2 == 0:
            count += 1
    return count

n=int(input())
result = count_even_divisors(n)
print(result)
```

```python
#BAI20
n = int(input("Nhập vào một số nguyên dương n: "))
if n <= 0:
    print("Vui lòng nhập một số nguyên dương.")
else:
    is_power_of_two = (n & (n - 1) == 0) and n != 0
    if is_power_of_two:
        print(f"{n} là số dạng 2^k.")
    else:
        print(f"{n} không phải là số dạng 2^k.")
```

```python
#BAI21
n_str = input("Nhập vào số nguyên dương n: ")
if not n_str.isdigit() or int(n_str) <= 0:
    print("Vui lòng nhập một số nguyên dương hợp lệ.")
else:
    total_sum = 0
    for digit in n_str:
        total_sum += int(digit)
    print(f"Tổng các chữ số của {n_str} là: {total_sum}")
```

```python
#BAI22
n_str = input("Nhập vào số nguyên dương n: ")
if not n_str.isdigit() or int(n_str) <= 0:
    print("Vui lòng nhập một số nguyên dương hợp lệ.")
else:
    count_even = 0
    count_odd = 0
    for digit_char in n_str:
        digit = int(digit_char)
        if digit % 2 == 0:
            count_even += 1
        else:
            count_odd += 1
    print(f"Số chữ số chẵn: {count_even}")
    print(f"Số chữ số lẻ: {count_odd}")
```

```python
#BAI23
n = int(input("Nhập n: "))
if n <= 0:
    print("Vui lòng nhập một số nguyên dương.")
else:
    k = 0
    S_k = 0
    while S_k + (k + 1) < n:
        k += 1
        S_k += k
    print(f"Giá trị k lớn nhất sao cho S(k) < n là: {k}")
```
