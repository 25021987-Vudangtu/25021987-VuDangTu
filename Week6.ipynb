```python
#BAI9
dic1=list(input().strip("{}").split(", "))
dic2=list(input().strip("{}").split(", "))
dic={}
for i in dic1:
  x,y=i.split(":")
  x=x.strip("'")
  y=int(y)
  if x not in dic:
    dic[x]=y
  else:
    dic[x]+=y

for j in dic2:
  x,p=j.split(":")
  x=x.strip("'")
  p=int(p)
  if x not in dic:
    dic[x]=p
  else: dic[x]+=p

print(dic)


```

```python
#BAI10
lst=list(map(int,input().split()))
lst1=[]
k=int(input())
for i in range(len(lst)):
  for j in range(i+1,len(lst)):
    if lst[i]+lst[j]==k:
      lst1.append((lst[i],lst[j]))
print(list(set(lst1)))

```

```python
#BAI21
def giai_quyet_bai_toan(chuoi_dau_vao):
    cac_tu = chuoi_dau_vao.split()
    vi_tri_tu = {}
    for chi_muc, tu in enumerate(cac_tu):
        if tu not in vi_tri_tu:
            vi_tri_tu[tu] = []
        vi_tri_tu[tu].append(chi_muc)
    for tu in vi_tri_tu:
        vi_tri_tu[tu] = tuple(vi_tri_tu[tu])
    return vi_tri_tu
```
