```python
#BAI1
n=list(map(int,input().split()))
x=[]
for i in n:
  if i not in x:
    x.append(i)
print(x)
```

```python
#BAI2
n=list(map(int,input().split()))
x=[]
s=0
for i in n:
  s+=i
  x.append(s)
print(x)
```

```python
#BAI3
n=list(map(int,input().split()))
x=[]
k=int(input())
for i in range(k,len(n)):
  x.append(n[i])
for i in range(0,k):
  x.append(n[i])
x=tuple(x)
print(x)
```

```python
#BAI4
lst=list(input().split())
dic={}
for i in lst:
  x,y=i.split(":")
  if x not in dic:
    dic[x]=[y]
  else:
    dic[x].append(y)
print(dic)


```

```python
#BAI5
lst=list(map(int,input().split()))
dic={}
p=0
n=0
z=0
for i in lst:
  if i >0:
    p+=1
  elif i<0: n+=1
  else: z+=1
dic["positives"]=p
dic["negatives"]=n
dic["zeros"]=z
print(dic)
```
